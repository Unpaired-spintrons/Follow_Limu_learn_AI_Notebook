## 线性回归算法的从零实现

### tor.matmul

```python
X = tor.tensor([1,2])
Y = tor.tensor([[1],[2]])

Z = tor.matmul(X, Y)
Z.shape
```

- 函数名称：tor.matmul 【专门强调torch是因为其他知名python库如numpy中也有matmul】
- 函数作用：用于进行线性代数矩阵相乘
- 函数输入：两个相乘矩阵(torch_tensor)注意前后
- 函数输出：相乘积后的矩阵(torch_tensor)
- 函数依赖：无

### d2l_tor.set_figsize()

```python
d2l_tor.set_figsize() # 创建一个图(类似于matlab需要一个figure)
```

- 函数名称：set_figsize()
- 函数作用：创建一个图表
- 函数输入：无
- 函数输出：创建一个图标账户
- 函数依赖：无 (这个是李沐的mxnet系代码)

### d2l_tor.plt.scatter(x_label, y_label, 图标号);

```python
y_label = tor.tensor([1,2,3,4,5,6])
x_label = tor.tensor([1,2,3,4,5,6])

d2l_tor.plt.scatter(x_label.detach().numpy(),
                    y_label.detach().numpy(),1);
# 上面的numpy转换是用来进行数据转换的
```

- 函数名称：d2l_tor.plt.scatter(x_label, y_label, 图标号);
- 函数作用：用来进行画图
- 函数输入：x_label,y_label,图标号【这些label是需要进行numpy转化的】
- 函数输出：输出一张图
- 函数依赖：无
